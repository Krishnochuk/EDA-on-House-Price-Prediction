# Numerical features are of two types

2. Discrete Variable( they have fixed set of integers)



```python
# Numerical features are of two types
## 1. Continous variable  2. Discrete Variable( they have fixed set of integers)


discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['id]']]
discrete_feature
```

```python
dataset[discrete_feature].head(3)
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    
    plt.ylabel('Sales Price')
    plt.show()
    
```

```python
continous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id']]
continous_feature
```

```python
for feature in continous_feature:
    data=dataset.copy()
    
    data[feature].hist()
    plt.show()
    
```

# Log-Normal ditribution

```python
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.show()
```

# Outliers

```python
#boxplot works with numerical feature well with continous feature
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass                                  
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
